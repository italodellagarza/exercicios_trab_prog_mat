**IMPLEMENTAO EM JULIA**

Importando JuMP

```julia
using JuMP
```

Importando Gurobi

```julia
using Gurobi
```

Criando o modelo

```julia
m = Model(with_optimizer(Gurobi.Optimizer))
```

Criando as variveis de deciso

```julia
@variable(m, q1 >= 0)
@variable(m, q2 >= 0)
@variable(m, q3 >= 0)
```

Criando a funo-objetivo

```julia
@objective(m, Max, 324.9*q1 + 296.4q2 + 413.2q3)
```

Definindo as restries

```julia
@constraints(m, begin
    2*q1 <= 40
    2*q2 <= 30
    2*q3 <= 10
    2*q1 + 2*q2 + 2*q3 <= 60
    q1 + (3*q2/4) + (5*q3/3) <= 40
    5.1*q1 + 3.6*q2 + 6.8*q3 <=200
    end)
```

Imprimindo o modelo

```julia
m
```

Otimizando a funo

```julia
start = time()
optimize!(m)
println("Tempo: $(time() - start)s")
```

Imprimindo os valores das variveis e o valor timo da funo objetivo

```julia
println("Valor timo: q1 = ", JuMP.value(q1), ", q2 = ", JuMP.value(q2), ", q3 = ", JuMP.value(q3))
println("Funo tima: ", JuMP.objective_value(m))
```

Ou seja, o melhor negcio a ser feito  comprar 20 Boeings 717, 5 Boeings 737-500 e 5 MD-11, lucrando com isso 10046 milhes de dlares.
