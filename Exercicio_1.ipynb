**5 - Empresa de Viação Aérea (Goldbarg & Luna, 2005).** A Empresa de Viação Aérea Brasileira está estudando a compra de três tipos de aviões: Boeing 717 para as pontes aéreas a curta distância, Boeing 737-500 para vôos domésticos e internacionais de média distância e MD-11 para vôos internacionais de longa distância. | BOEING 717 | 5,1 | 330 | 30 |

**IMPLEMENTAÇÃO EM JULIA**

Importando JuMP

```julia
using JuMP
```

Importando Gurobi

```julia
using Gurobi
```

Criando o modelo

```julia
m = Model(with_optimizer(Gurobi.Optimizer))
```

Criando as variáveis de decisão

```julia
@variable(m, q1 >= 0)
@variable(m, q2 >= 0)
@variable(m, q3 >= 0)
```

Criando a função-objetivo

```julia
@objective(m, Max, 324.9*q1 + 296.4q2 + 413.2q3)
```

Definindo as restrições

```julia
@constraints(m, begin
    2*q1 <= 40
    2*q2 <= 30
    2*q3 <= 10
    2*q1 + 2*q2 + 2*q3 <= 60
    q1 + (3*q2/4) + (5*q3/3) <= 40
    5.1*q1 + 3.6*q2 + 6.8*q3 <=200
    end)
```

Imprimindo o modelo

```julia
m
```

Otimizando a função

```julia
start = time()
optimize!(m)
println("Tempo: $(time() - start)s")
```

Imprimindo os valores das variáveis e o valor ótimo da função objetivo

```julia
println("Valor ótimo: q1 = ", JuMP.value(q1), ", q2 = ", JuMP.value(q2), ", q3 = ", JuMP.value(q3))
println("Função ótima: ", JuMP.objective_value(m))
```

Ou seja, o melhor negócio a ser feito é comprar 20 Boeings 717, 5 Boeings 737-500 e 5 MD-11, lucrando com isso 10046 milhões de dólares.
